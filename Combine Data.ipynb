```python
import pandas as pd
import os
import re
```

```python
df = pd.DataFrame()
```

```python
path = '../../Documents/BMKG_Kota_Bandar_Lampung'
os.listdir(path)
```

```python
for folder in os.listdir(path):
    for file in os.listdir(folder):
        if file.endswith('.xlsx'):
            file_path = f"{path}/{folder}/{file}"
            df = df.append(pd.read_excel(file_path))
```

```python
df.head()
```

```python
df = df.rename(columns = {
    'Unnamed: 0': 'Tanggal',
    'ID WMO': 'Tn',
    ':  96293': 'Tx',
    'Unnamed: 3': 'Tavg',
    'Unnamed: 4': 'RH_avg',
    'Unnamed: 5': 'RR',
    'Unnamed: 6': 'ss',
    'Unnamed: 7': 'ff_x',
    'Unnamed: 8': 'ddd_x',
    'Unnamed: 9': 'ff_avg',
    'Unnamed: 10': 'ddd_car'
})
```

```python
df = df[df.Tanggal.apply(lambda x: bool(re.match(r'\d{2}-\d{2}-\d{4}', str(x))))]
```
